```python
import pandas as pd
```

```python
df1 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2014-to-01-01-2015.csv')
df2 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2015-to-01-01-2016.csv')
df3 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2016-to-01-01-2017.csv')
df4 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2017-to-01-01-2018.csv')
df5 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2018-to-01-01-2019.csv')
df6 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2019-to-01-01-2020.csv')
df7 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2020-to-01-01-2021.csv')
df8 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2021-to-01-01-2022.csv')
df9 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2022-to-01-01-2023.csv')
df10 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2023-to-01-01-2024.csv')
df11 = pd.read_csv('united_spirits_data_01_Jan_2014_-_20_Sep_2024/Quote-Equity-UNITDSPR-EQ-01-01-2024-to-20-09-2024.csv')
```

```python
df = pd.concat([df11, df10, df9, df8, df7, df6, df5, df4, df3, df2, df1])
df
```

```python
df['Date '] = pd.to_datetime(df['Date '], format='%d-%b-%Y')
df
```

```python
# Change column name to that of Yahoo Finance format
df = df[['Date ', 'OPEN ', 'HIGH ', 'LOW ', 'close ', 'VOLUME ']]

df.rename(columns= {'Date ':'Date', 
                    'OPEN ':'Open', 
                    'HIGH ':'High', 
                    'LOW ':'Low', 
                    'close ':'Close', 
                    'VOLUME ':'Volume'}, inplace=True)

# Remove commas from values
df = df.replace(',','', regex=True)

# Reverse series order
df = df.iloc[::-1]

# Reset index
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)

df['Date'] = df['Date'].dt.date
df.set_index('Date', inplace=True)
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')

df = df[:-1]
df.drop_duplicates(inplace=True)

df
```

```python
# Import any stock from NIFTY 50. The dates from the data will be used to filter dates from df.
import yfinance as yf

stock_data = yf.download('HDFCBANK.NS', start='2014-01-01', end='2024-09-20')
stock_data
```

```python
# Find common date values
stock_data.index.intersection(df.index)
```

```python
new_df = df.loc[stock_data.index.intersection(df.index)]
new_df
```

```python
new_df.to_csv('UNITDSPR.NS_2014-01-01_to_2024-09-19.csv')
```
